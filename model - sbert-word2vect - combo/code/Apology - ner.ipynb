```python
replace_email = []
for res in corpus:
    email_lst = re.findall(r'[\w.+-]+@[\w-]+\.[\w.-]+', res)
    for email in email_lst:
        res = res.replace(email, "EMAIL")
    replace_email.append(res)
```

```python
clean_corpus = []
for res in replace_email:
    clean_corpus.append(res.replace(".", ". ").replace(",", ", "))
    
```

```python
doc = ner.name(clean_corpus[0], language = "en_core_web_sm")
```

```python
text_label = [(X.text, X.label_) for X in doc]
```

```python
ner_corpus = []
for res in clean_corpus:
    doc = ner.name(res, language = "en_core_web_sm")
    text_label = [(X.text, X.label_) for X in doc]
    for tup in text_label:
        res = res.replace(tup[0], tup[1])
    ner_corpus.append(res)
    
    
```
